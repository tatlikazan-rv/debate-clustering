```python
import tomotopy as tp
import topic_modeling_utils
import utils
import nltk

nltk.download('stopwords')
corpus = topic_modeling_utils.tomotopy_corpus()
sentences= utils.data_lines()
```

```python
lda_model = tp.LDAModel(corpus=corpus, k=10)
lda_model.train(1000)
doc_inst = lda_model.make_doc(sentences)
lda_model.infer(doc_inst)

lda_model.save("LDA.bin")
# estimate distribution
for k in range(lda_model.k):
    print('Top 10 words of topic #{}'.format(k))
    print(lda_model.get_topic_words(k, top_n=10))
```
