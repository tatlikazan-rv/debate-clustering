```python

import utils
import topic_modeling_utils
import tomotopy as tp
import sklearn.decomposition as skd
%matplotlib inline
utils.FILE_NAME= "../data/Game_of_Thrones_Script.csv"
feature_names, matrix = topic_modeling_utils.tf_idf_vectorizer(utils.data_lines())
```

```python
lda_bow = skd.LatentDirichletAllocation(n_components=10, random_state=42)
doc_topic = lda_bow.fit_transform(matrix)
topic_modeling_utils.plot_cluster_metrics_sc(doc_topic)
```

```python
LSA_model = skd.TruncatedSVD(n_components=10, n_iter=10, random_state=42)
doc_topic = LSA_model.fit_transform(matrix)
topic_modeling_utils.plot_cluster_metrics_sc(doc_topic)
```

```python
nmf_model = skd.NMF(n_components=10, init="nndsvda", beta_loss="kullback-leibler", solver="mu", alpha=0.00005, max_iter=1000, l1_ratio=0.5)
doc_topic = nmf_model.fit_transform(matrix)
topic_modeling_utils.plot_cluster_metrics_sc(doc_topic)
```

```python
model = tp.LDAModel.load("LDA.bin")
words = [[model.get_topic_words(k, top_n=10)[i][0] for i in range(10)] for k in range(model.k)]
doc_topic = [doc.get_topic_dist() for doc in model.docs]
topic_modeling_utils.plot_cluster_metrics_sc(doc_topic)
```
